```python
import cv2
import numpy as np 
```

```python
img = cv2.imread('dahliya.jpg')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img[:,:,::-1]);
```

```python
crop = img[0:2000, 0:3000]
plt.imshow(crop[...,::-1]);
```

```python
# Generate a black bg.
black_bg = np.zeros((2000,3000), dtype=np.uint8)
black_bg = [black_bg, black_bg, black_bg]
black_bg = cv2.merge(black_bg)
plt.imshow(black_bg, cmap='gray');
```

```python
img_temp = img.copy()
img_temp[0:2000, 0:3000] = black_bg
plt.imshow(img_temp[...,::-1]);
```

```python
r, g, b = cv2.split(crop)
r = r + 20

cropped_ls = [r, g, b]
merged = cv2.merge(cropped_ls)
plt.imshow(merged[...,::-1]);
```

```python
red = np.ones((2000,3000), dtype=np.uint8)
red = 255 * red
green = np.zeros((2000,3000), dtype=np.uint8)
blue = np.zeros((2000,3000), dtype=np.uint8)

ls = [blue, green, red]

red_pallet = cv2.merge(ls)
plt.imshow(red_pallet[...,::-1]);
```

```python
reddened = cv2.add(crop, red_pallet)
plt.imshow(reddened[...,::-1])
```

```python
img[0:2000, 0:3000] = reddened
plt.imshow(img[...,::-1]);
```
